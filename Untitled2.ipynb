```python
import pandas as pd
from transformers import pipeline
import json

# Load Dataset from JSON
df = pd.read_json('/content/Dataset.json')

# Load HuggingFace Pre-trained Model for NER
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")

# Function to Mask Entities
def mask_entities(text):
    ner_results = ner_pipeline(text)
    for entity in ner_results:
        # Replace the entity with [MASK]
        text = text.replace(entity['word'], '[MASK]')
    return text

# Apply Masking to All Rows
df['masked_text'] = df['full_text'].apply(mask_entities)

# Display First 5 Rows with Masked Output
print(df[['full_text', 'masked_text']].head())

# Save Output to CSV (Downloadable File)
df.to_csv('/content/Masked_Data.csv', index=False)
print("✅ Masking Completed. Download File Below:")
```

```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load Masked Data
df = pd.read_csv('/content/Masked_Data.csv')

# Extract true and predicted labels
true_labels = []
pred_labels = []

for i in range(len(df)):
    true_labels.extend(eval(df['labels'][i]))
    pred_labels.extend(['O' if '<mask>' in t else t for t in eval(df['tokens'][i])])

# Generate Classification Report
report = classification_report(true_labels, pred_labels, output_dict=True)
report_df = pd.DataFrame(report).transpose()
report_df.to_csv('/content/classification_report.csv')

# Generate Confusion Matrix
conf_matrix = confusion_matrix(true_labels, pred_labels, labels=list(set(true_labels)))
plt.figure(figsize=(12, 8))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.savefig('/content/Confusion_Matrix.png')

print("✅ Evaluation Complete!")
```
